```python
import pandas as pd
import sklearn as sl
import numpy as np

from sklearn.model_selection import train_test_split
```

## Caricamento del dataset

Carichiamo il dataset appointment_data.csv da utilizzare per addestrare e testare il modello che andremo a creare.

```python
# Il file csv del dataset viene scaricato dalla repository git
url = 'appointment_data.csv'

app_data = pd.read_csv(url, na_values=np.nan)
```

```python
app_data.head(10)
```

## Split del dataset

Dividiamo il dataset in due parti, una per l'operazione di train e una per quella di test del modello.

```python
# Dividiamo il dataset in due parti:
# una per il test e una per il train.
train, test = train_test_split(app_data, test_size=.2)
```

Salviamo i due dataset appena creati in due file distinti.

```python
train.to_csv('train.csv', index=False)
test.to_csv('test.csv', index=False)
```
