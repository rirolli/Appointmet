```python
import pandas as pd
import sklearn as sl
import numpy as np

from sklearn.model_selection import train_test_split
```

## Caricamento del dataset completo

Caricamento del dataset appointment_data.csv.

```python
url = 'appointment_data.csv'

app_data = pd.read_csv(url, na_values=np.nan)
```

```python
app_data.head(15)
```

## Split del dataset

Dividiamo il dataset in tre parti: Train, Validation e Test

```python
# Dividiamo il dataset in tre parti: Train,
# Validation e Test.
train, test = train_test_split(app_data, test_size=.2)
train, validation = train_test_split(train, test_size=.2)
```

Saviamo i dataset appena creati in file csv

```python
train.to_csv('train.csv', index=False)
validation.to_csv('validation.csv', index=False)
test.to_csv('test.csv', index=False)
```
